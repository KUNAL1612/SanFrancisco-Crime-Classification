```python
from matplotlib import pyplot as plt
import numpy as np


crime_categories = crime_data['Category'].value_counts().index



# get a 39-d outer array ,each of whose ith element is an array, 
#containing the instances of features,for that ith crime,ie the ith element of crime_categories array.
def get_crime_instances_for_feature(feature,crime_categories,crime_data):
    
    crime_instances = []
    
    for i in crime_categories:
       
        crime_instances.append(crime_data.loc[crime_data['Category'] == i][feature])
    
    return crime_instances


# plot a stacked histogram.
def plot_stacked_hist_for_feature(feature,crime_categories,crime_data):
    
    bins = np.sort(crime_data[feature].unique())
    
    print(bins)
    
    
    crime_instances = get_crime_instances_for_feature(feature,crime_categories,crime_data)   
    #print(crime_instances)

    plt.figure()

    plt.hist(crime_instances, bins, stacked=True,rwidth=10,label = crime_categories)
    #plt.legend("title", prop=fontP)
    plt.legend(loc=9, bbox_to_anchor=(0.5, -0.1), ncol=2)

    plt.show()

feature = 'Day'
plot_stacked_hist_for_feature(feature,crime_categories,crime_data)
```

```python
from sklearn.preprocessing import LabelBinarizer

encoder = LabelBinarizer()


def one_hot_encoding_features(crimes):
    

    #day_of_week = crimes['DayOfWeek'].unique()

    #day_of_week_1hot = encoder.fit_transform(day_of_week)
    pddistrict_onehot = pd.get_dummies(crimes['PdDistrict'])
    pddistrict_onehot["Index"] = np.arange(878049)

    #print(onehot)

    crimes = crimes.drop('PdDistrict',axis=1)
    crimes = pd.merge(crimes,pddistrict_onehot,on='Index')
    
    #pd_district = crimes['PdDistrict'].unique()

    #pd_district_1hot = encoder.fit_transform(pd_district)
    
    print('Encoding done !!!')
    return (crimes,pddistrict_onehot)
    
    
    
    







def transform_cyclical_features(crimes):
        
    crimes['hourfloat'] = crimes['Hour'] + crimes['Minutes']/60.0
    
    
    crimes['hourfloat_sine']= np.sin(crimes.hourfloat*(2.*np.pi/24.))
    crimes['hourfloat_cosine']= np.cos(crimes.hourfloat*(2.*np.pi/24.))
    
    crimes['month_sine']= np.sin((crimes['Month'] - 1)*(2.*np.pi/12.))
    crimes['month_cosine']= np.cos((crimes['Month'] - 1)*(2.*np.pi/12.))
    
    crimes['day_sine']= np.sin((crimes['Day'] - 1)*(2.*np.pi/31.))
    crimes['day_cosine']= np.cos((crimes['Day'] - 1)*(2.*np.pi/31.))
    
    
    print('Cyclical features transformed into sine and cosine!!!')

    
    
crimes,pddistrict_onehot = one_hot_encoding_features(crimes)   

transform_cyclical_features(crimes)

crimes.info()
```
